```python
# We can acces to multiple columns (a smaller dataframe)
online_retail_df[['InvoiceNo', 'Quantity', 'UnitPrice']].head()
# This is a dataframe
```

```python
# We can acces to just one column as a dataframe
wanted_column = ['Country']
online_retail_df[wanted_column].head() # This is a dataframe
```

```python
# We can group by a trait and obtain the aggregation (sum, max, min, mean, std, etc. )
online_retail_df_by_country = online_retail_df.groupby("Country")
online_retail_df_by_country.sum()
```

```python
# We can group by a trait and obtain more aggregations
# (sum, max, min, mean, std, etc. )
online_retail_df_by_country = online_retail_df.groupby("Country")
online_retail_df_by_country.agg({'Quantity':'sum',
                                 'UnitPrice':'mean', 
                                 'CustomerID':'count',
                                 'Description':'last'})
```

```python
# Groupby can be done with multiple columns
online_retail_df_by_invoice_country = online_retail_df.groupby(['InvoiceNo', 'Country']).sum()
online_retail_df_by_invoice_country
```

```python
online_retail_df_by_invoice_country.reset_index()
```

```python
# Acess individual rows
online_retail_df_by_invoice_country.iloc[9] # locate the position by INDEX
```

```python
online_retail_df_by_invoice_country.iloc[1:10] # Locate many posotions by INDEX
```

```python
# Apply filters of information
'''
& and
| or
~ not
'''
online_retail_df[(online_retail_df.Quantity < 0) & (online_retail_df.CustomerID == 15311)].head()
```

```python
# Apply operations
online_retail_df.eval("TotalPrice = Quantity * UnitPrice")
```

```python
# Lets see what happens
online_retail_df.head()
```

```python
# We can even create more stuff on the dataframe
online_retail_df['AbsTotalPrice'] = abs(online_retail_df.Quantity*online_retail_df.UnitPrice)
online_retail_df.head()
```

```python
# Create other columns with the conditions of the dataframe
online_retail_df.loc[online_retail_df.Quantity < 0,'IsQuantityNeg'] = 1
```

```python
online_retail_df.head()
```

```python
online_retail_df.IsQuantityNeg.fillna(0, inplace=True)
online_retail_df.head()
```

```python
online_retail_df.IsQuantityNeg.mean()*100
```
