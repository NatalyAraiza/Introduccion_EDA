## Leer datos en `pandas`




Una forma popular de guardar datos es en archivos .csv y luego leer esos archivos en Python usando Pandas. Los archivos .csv son archivos de texto plano (o sin formato) que almacenan los datos en forma de tablas, separando los valores de las columnas usando comas.

Para leer un archivo .csv en `pandas` primero debes importar la libreria, (por convención, se usa el alias `pd`), luego llamas a la función `read_csv()`:

~~~python
import pandas as pd
pd.read_csv("path/to/file.csv")
~~~
